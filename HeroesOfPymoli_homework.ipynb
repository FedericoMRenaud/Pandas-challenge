```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_data_to_load = "purchase_data.csv"

# Read file
purchase_data = pd.read_csv(purchase_data_to_load)
purchase_data.head()
```

```python
total_players=len(purchase_data ["SN"].value_counts())
totalplayers=pd.DataFrame({"Total players":[total_players]})
totalplayers
```

```python
unique_items = len(purchase_data["Item ID"].value_counts())
unique_items
print("Unique items: {}".format(unique_items))

purchases = len(purchase_data["Purchase ID"].value_counts())
prices = purchase_data["Price"].sum()
average_price = prices/purchases
print("Average price: {}".format(average_price))

print("Number of purchases: {}".format(purchases))

prices2=("${}".format(prices))
print(prices2)
```

```python
totals=pd.DataFrame({"Unique items":[unique_items],"Average price": [average_price], "Number of purchases":[purchases],"Total revenue": [prices2]})

totals.head()
```
